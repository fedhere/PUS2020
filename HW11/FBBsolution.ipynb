## Motivation
One of the things that make NYC so interesting to study is its diversity on many axes: ethnic, economic, cultural. Neighborhoods tend to be segregated and clustered on most socioeconomic characteristics, however these characteristics do evolve in time. 
The goal is to identify NYC neighborhoods that are homogeneous by clustering their socioeconomic characteristics in 2000, and in 2010, and find the neighborhoods that transition between clusters, indicating that they evolve differently with respect to their cluster peers (e.g. gentrification).

Note: in this project socioeconomic distributions will be used and aggregated at various levels. As data gets aggregated the distributions will change due to loss of datapoints  in the aggregation. We will explore and describe these changes and note that they may generate artifacts and false positives or negatives.  But we will continue this exploration noting that its robustness to these changed needs to be verified

```python
# Load data to pandas dataframes.
df00 = pd.read_csv('data/zbp00totals.txt', 
                   usecols=['ZIP', 'NAME', 'EST'])
df10 = pd.read_csv('data/zbp10totals.txt', 
                   usecols=['zip', 'name', 'est'])

# Rename columns in 2000 data.
df00.rename(columns={'NAME': 'name', 'ZIP': 'zip', 'EST': 'est'}, 
            inplace=True)

# Remove data outside of NY.
df00 = df00[df00.name.str[-2:] == 'NY']
df10 = df10[df10.name.str[-2:] == 'NY']

# Convert 'zip' to string.
df00['zip'] = df00['zip'].map(lambda x: "%05d"%x)
df10['zip'] = df10['zip'].map(lambda x: "%05d"%x)

# Set zip to index.
df00.set_index('zip', inplace=True)
df10.set_index('zip', inplace=True)

# Join business data on zip index.
biz = df00.join(df10, how='outer', lsuffix='2000', rsuffix='2010')

# Clean up dataframe columns.
biz.drop('name2010', axis=1, inplace=True)
biz.rename(columns={'name2000': 'name'}, inplace=True)
biz['est2000'] = biz.est2000.astype(float)

biz.sort_index().head(2), biz.sort_index().tail(2)
```

# AGGREGATING CENSUS TRACTS AT THE ZIPCODE LEVEL

## Merging census tract and zipcode geodataframes
The census tracts need to be aggregated at the zipcode level since the business data is available at that coarser granularity.
However there are two issues in this: 

1) how to aggregate medians. Median values do not strictly aggregate, but this is not important in our case since we are not looking for an unbiased estimator, but only of a metric that can be compared at the zipcode level. So as long as I am consistent in the aggregation I can compare zipcode aggregated values. I will aggregate the value available at the census tract level by taking their mean within the zipcode.

2) The census tracts do not have to be contained neatly within zipcodes, since the census tracts evolve in the attempt to sample that are collect socioeconomic homogeneous and even in population. A weighted average of the census tracts that accunts for the intersection of the census tract with the zipcode would be sensible. This could be obtained with a spatial joint that finds the overlapping area of a zipcode with a censust tract in frations of the zipcode area. A more straightforward approach is to find the zipcode that contains the center of the census tract and assign the tract to this zipcode. Given the small relative size of census tract and zipcode this approach is sufficiently accurate to not to contribute as a leading source of errors. This can be accomplished in a for-loop.

# CLUSTERING
The goal is to identify neighborhoods that have a qualitatively different evolution in the 10 year span we are considering, with respect to neighborhood that start in  similar socioeconomic conditions in 2000. We will do this by first clustering similar neighborhoods with each others, then seeing the evolution of those clusters. We could simply look at points in the periphery of the cluster in the final state of the cluster that were not "outliers" in the initial state, but we can instead exploit the properties of clustering and see what fits together, and what splits off. The added advantage is that the neighborhoods not only have to evolve differently from the present cluster, but they have to evolve to be more similar to other NYC neighborhoods. Gentrification fits this pattern: a diverse low socioeconomic neighborhoods that becomes wealthier, whiter, and possibly younger (or maybe jut less diversified in age). 


```python
vals00 = skl.preprocessing.scale(zipBoth.dropna()[[feat + '_00' for feat in allfeatures]])
vals10 = skl.preprocessing.scale(zipBoth.dropna()[[feat + '_10' for feat in allfeatures]])
```

```python
from sklearn.metrics import silhouette_score

pl.figure(figsize=(10,10))
for i in range(5):
    s = []
    for n_clusters in range(2,10):
        kmeans = KMeans(n_clusters=n_clusters, init='k-means++')
        kmeans.fit(vals00)

        labels = kmeans.labels_
        centroids = kmeans.cluster_centers_

        s.append(silhouette_score(vals00, labels, metric='euclidean'))

    pl.plot(range(2,10), s)
pl.ylabel("Silouette", fontsize = 18)
pl.xlabel("number of clusters", fontsize = 18)
pl.title("Silouette for K-means cell's behaviour", fontsize = 18)
```
